```python
% matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
url_set = "./grillos.csv"
df = pd.read_csv(url_set,header = 0)
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.iloc[:,:1].values
```

```python
Y = df.iloc[:,1:].values
```

```python
fig = plt.figure()
plt.grid(False)
fig = plt.scatter(X, Y, color='green')
plt.suptitle("Chirridos vs Temperatura")

plt.xlabel('Temperatura') 
plt.ylabel('Chirridos')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# create training and testing vars
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
print( X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
model = regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```

```python
accuracy = r2_score(y_test, y_pred)
print("PrecisiÃ³n de predicciÃ³n cruzada:", accuracy)
```

```python
# Plot outputs
fig = plt.figure()
X_test = X_test
plt.scatter(X_test, y_test,  color='green')
plt.scatter(X_train, y_train,  color='red')
plt.plot(X_test, y_pred, color='blue', linewidth=5)
plt.grid()

plt.xticks(())
plt.yticks(())
plt.suptitle("Chirridos vs Temperatura")
plt.xlabel('Temperatura') 
plt.ylabel('Chirridos')

plt.show()
```
